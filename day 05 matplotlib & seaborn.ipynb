## pie chart

```python
langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
my_explode = (0.0, 0.1, 0.1)

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = my_explode, shadow =  True)
plt.title("Pie chart")
plt.show()
```
